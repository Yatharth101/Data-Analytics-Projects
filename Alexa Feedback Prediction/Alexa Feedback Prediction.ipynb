## Word pre-processing

```python
from nltk import word_tokenize
```

```python
data = df["verified_reviews"]
print(data)
```

```python
from nltk.corpus import stopwords
sw = stopwords.words('english')
print(sw)
```

```python
import re
```

```python
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()
msg = []
for i in data:
    t = re.sub('[^A-Za-z0-9]', ' ', i)       # Removing Punctuations
    t = t.lower()                            # Turning all to lower case
    t = word_tokenize(t)                     # Word tokenization
    t = [i for i in t if t not in sw]        # Removing stopwords
    t = [lm.lemmatize(i) for i in t]         # Lemmatization
    t = " ".join(t)
    msg.append(t)
print(msg)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
cv_res = cv.fit_transform(msg).toarray()
print(cv_res)
print(cv.get_feature_names())
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df["feedback"]
x_train, x_test, y_train, y_test = train_test_split(cv_res, y, test_size = 0.25, random_state = 42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
m1 = KNeighborsClassifier()
m1.fit(x_train, y_train)
```

```python
y_pred = m1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def eval_model(ypred, ytest):
    cm = confusion_matrix(ytest, ypred)
    print(cm)
    print(classification_report(ytest, ypred))

def mscore(model):
    print("Training Score", model.score(x_train, y_train))
    print("Testing Score", model.score(x_test, y_test))
```
